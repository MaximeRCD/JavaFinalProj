# Making good formatted fake Dataframe

```python
import csv
import pandas as pd
from faker import Faker
import datetime
import random
from faker.providers import DynamicProvider


def fake_data_generation(records):
    fake = Faker('fr_FR')
    code_soin = [""]
    
    employee = []
    
    for i in range(records):
        first_name = fake.first_name()
        last_name = fake.last_name()



        employee.append({
                "ssn": fake.ssn(),
                "prenom": first_name,
                "nom": last_name,
                "date_naissance" : fake.date_of_birth(minimum_age=18, maximum_age=90).strftime("%Y-%m-%d"),
                "email": str.lower(f"{first_name}.{last_name}{random.randint(0,999)}@{fake.domain_name()}"),
                "numero_telephone" : fake.phone_number(),
                "id_remboursement" : random.randint(1000,9999),
                "code_soin" : list(code_maladie.keys())[random.randint(0, len(code_maladie.keys())-1)],
                "montant_remboursement" : random.randint(0,1000)/random.randint(10,20)
                })
        
    return employee


df = pd.DataFrame(fake_data_generation(10))
```

# Making bad formatted fake Dataframe

```python
import datetime
import random
from faker.providers import DynamicProvider

skill_provider = DynamicProvider(
     provider_name="skills",
     elements=["Python", "Pandas", "Linux", "SQL", "Data Mining"],
)


def fake_data_generation(records):
    fake = Faker('fr_FR')
    currencies = ['£','€','$'] 
    employee = []
    
    fake.add_provider(skill_provider)

    for i in range(records):
        first_name = fake.first_name()
        last_name = fake.last_name()



        employee.append({
                "ssn": fake.ssn(),
                "prenom": first_name,
                "nom": last_name,
                "date_naissance" : fake.date_of_birth(minimum_age=18, maximum_age=90).strftime("%Y-%m-%d"),
                "email": str.lower(f"{first_name}.{last_name}{random.randint(0,999)}@{fake.domain_name()}"),
                "numero_telephone" : fake.phone_number(),
                "id_remboursement" : random.randint(1000,9999),
                "code_soin" : list(code_maladie.keys())[random.randint(0, len(code_maladie.keys())-1)],
                "montant_remboursement" : str(random.randint(0,1000)/random.randint(10,20)) + random.choice(currencies) ## Rajoute le symbol euros à la fin qui ne convient pas au type float attendu
                })
        
    return employee


df = pd.DataFrame(fake_data_generation(20))
```
